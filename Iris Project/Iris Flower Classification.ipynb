```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
data['Species']=le.fit_transform(data['Species'])
data.head()
```

```python
from sklearn.model_selection import train_test_split
#train  -70
#test  -30
X=data.drop(columns=['Species'])
Y=data['Species']
x_train,x_test,y_train,y_test=train_test_split(X, Y, test_size=0.30)
```

```python
#logistic Regression
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
#model Training
model.fit(x_train,y_train)
```

```python
# print metric to get performancce
print("Accuracy :", model.score(x_test,y_test)*100)
```

```python
import tkinter as tk
from tkinter import ttk
import joblib
import pandas as pd  # Make sure to import pandas

class IrisFlowerClassification:
    def __init__(self, master):
        self.master = master
        master.title("Iris_Flower_Classification")
        master.geometry("400x400")
        
        # Load model once during initialization
        try:
            self.model = joblib.load('Iris_Flower_Classification')
            self.model_loaded = True
        except Exception as e:
            self.model_loaded = False
            print(f"Error loading model: {e}")
        
        # Mapping numerical predictions to species names (if needed)
        self.species_map = {
            0: "Iris-setosa",
            1: "Iris-versicolor",
            2: "Iris-virginica"
        }
        
        self.create_widgets()
    
    def create_widgets(self):
        # Main frame for better organization
        main_frame = ttk.Frame(self.master, padding="10")
        main_frame.pack(fill=tk.BOTH, expand=True)
        
        # Title
        ttk.Label(main_frame, text="Iris Flower Classification", 
                 background="black", foreground="white").pack(fill=tk.X)
        
        # Input fields frame
        input_frame = ttk.Frame(main_frame)
        input_frame.pack(pady=10, fill=tk.X)
        
        # Fields configuration
        self.fields = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
        self.entries = {}
        
        for i, field in enumerate(self.fields):
            # Create label
            ttk.Label(input_frame, text=field).grid(row=i, column=0, sticky=tk.W, padx=5, pady=2)
            
            # Create entry with validation
            entry = ttk.Entry(input_frame)
            entry.grid(row=i, column=1, sticky=tk.EW, padx=5, pady=2)
            self.entries[field] = entry
            
            # Add validation for numerical fields
            entry.config(validate="key", 
                       validatecommand=(input_frame.register(self.validate_number), '%P'))
        
        # Configure column weights
        input_frame.columnconfigure(1, weight=1)
        
        # Button frame
        button_frame = ttk.Frame(main_frame)
        button_frame.pack(pady=10)
        
        # Predict button
        ttk.Button(button_frame, text="Predict", command=self.predict).pack(side=tk.LEFT, padx=5)
        
        # Clear button
        ttk.Button(button_frame, text="Clear", command=self.clear_fields).pack(side=tk.LEFT, padx=5)
        
        # Result label
        self.result_label = ttk.Label(main_frame, text="", font=('Arial', 10, 'bold'))
        self.result_label.pack(pady=10)
        
        # Status label
        self.status_label = ttk.Label(main_frame, text="")
        self.status_label.pack()
        
        if not self.model_loaded:
            self.status_label.config(text="Error: Model not loaded!", foreground="red")
    
    def validate_number(self, value):
        """Validate that the input is a valid number"""
        if value == "":
            return True
        try:
            float(value)
            return True
        except ValueError:
            return False
    
    def clear_fields(self):
        """Clear all input fields"""
        for entry in self.entries.values():
            entry.delete(0, tk.END)
        self.result_label.config(text="")
    
    def predict(self):
        """Make prediction based on input values"""
        if not self.model_loaded:
            self.result_label.config(text="Model not available", foreground="red")
            return
            
        try:
            input_data = []
            for field in self.fields:
                value = self.entries[field].get()
                if not value:
                    self.result_label.config(text=f"Please enter {field}", foreground="red")
                    return
                input_data.append(float(value))
            
            # Convert to DataFrame (ensure column names match training data)
            input_df = pd.DataFrame([input_data], columns=self.fields)
            
            # Make prediction
            prediction = self.model.predict(input_df)[0]
            
            # Map numerical prediction to species name (if needed)
            predicted_species = self.species_map.get(prediction, f"Unknown ({prediction})")
            
            self.result_label.config(text=f"Predicted Species: {predicted_species}", foreground="green")
        
        except Exception as e:
            self.result_label.config(text=f"Error: {str(e)}", foreground="red")

def main():
    root = tk.Tk()
    app = IrisFlowerClassification(root)
    root.mainloop()

if __name__ == "__main__":
    main()
```
